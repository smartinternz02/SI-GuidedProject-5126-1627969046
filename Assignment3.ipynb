```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
iris = sns.load_dataset("iris")
```

```python
plt.scatter(iris['petal_width'],iris['petal_length'],color="orange")
plt.title('Iris Dataset of Petal')
plt.xlabel('Petal Width')
plt.ylabel('Petal Length')
plt.show()
```

```python
sns.scatterplot(x="sepal_width",y="sepal_length",data=iris)
```

```python
plt.bar(iris['petal_width'],iris['petal_length'], color = "red")
plt.xlabel("Petal Width")
plt.ylabel("Petal Length")
plt.title("Petal Width vs Petal Length")
plt.show()
```

```python
plt.subplot(2,2,1)#(rows,columns , division number)
plt.plot(iris['sepal_width'],iris['sepal_length'] , color = "blue")
plt.subplot(2,2,2)#(rows,columns , division number)
plt.bar(iris['sepal_width'],iris['sepal_length'] , color = "green")
plt.subplot(2,2,3)#(rows,columns , division number)
plt.scatter(iris['sepal_width'],iris['sepal_length'] , color = "purple")
plt.show()
```
